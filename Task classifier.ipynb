```python
from langchain.chat_models import ChatOpenAI
from kor import create_extraction_chain, Object, Text, Number

llm = ChatOpenAI(
    model_name="gpt-3.5-turbo",
    temperature=0,
    max_tokens=2000,
    frequency_penalty=0,
    presence_penalty=0,
    top_p=1.0,
)
schema = Object(
  id="task",
  description=(
      "user is writing down tasks to do"
  ),
  attributes=[
      Text(
          id="action",
          description="User wants to extract the action item from the task",
          examples=[],
          many=True,
      ),
      Text(
          id="category",
          description="User wants to categorize the task into School if it is related to academics, Personal if it is related to self growth, and the rest under Household",
          # user wants to categorize the task into school, personal, and household according to the nature of the task
          examples=[],
          many=True,
      ),
    ],
  many=False,
)
```

```python
chain = create_extraction_chain(llm, schema, encoder_or_encoder_class='json')
chain.predict_and_parse(text="remind me to buy eggs")['data']
```
